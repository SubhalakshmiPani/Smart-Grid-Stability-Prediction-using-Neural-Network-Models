```python
# Detect outliers using IQR
Q1 = data.iloc[:,:13].quantile(0.25)
Q3 = data.iloc[:,:13].quantile(0.75)
IQR = Q3 - Q1

# Define outlier bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter outliers
outliers = ((data[numerical_columns] < lower_bound) | (data[numerical_columns] > upper_bound)).sum()
print("Outliers detected:\n", outliers)
```

Train Test Splitting
=

```python
train=data.iloc[:47999,:]
test=data.iloc[48000:,:]
X_train=train.iloc[:,:12]
y_train=train.iloc[:,12]
X_test=test.iloc[:,:12]
y_test=test.iloc[:,12]
```

Linear Regression Model
=

```python
# Import required libraries
import pandas as pd
# from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score

# Standardize the features (optional but recommended for regression)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Initialize the Linear Regression model
model = LinearRegression()

# Train the model on the training data
model.fit(X_train, y_train)

# Make predictions on the test data
y_pred = model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Model Performance:")
print(f"Mean Squared Error (MSE): {mse:.4f}")
print(f"R-squared (R²): {r2:.4f}")
```

```python
import numpy as np
y_pred=np.where(y_pred>0.5,1,0)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

# Example: Replace y_test with the true labels and y_pred_binary with your predicted binary labels
# Compute confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Print confusion matrix
print("Confusion Matrix:")
print(cm)

# Print additional metrics
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.4f}")
```

All comparision
=

```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC  # For SVM with RBF kernel
from sklearn.metrics import accuracy_score

# Initialize the models
models = {
    'Logistic Regression': LogisticRegression(),
    'Decision Tree': DecisionTreeClassifier(random_state=42),
    'Random Forest': RandomForestClassifier(n_estimators=100, random_state=42),
    'KNN': KNeighborsClassifier(n_neighbors=5),
    'SVM (RBF)': SVC(kernel='rbf', random_state=42)  # SVM with RBF kernel
}

# List to store model names and their accuracy
model_names = []
model_accuracies = []

# Train and evaluate each model
for model_name, model in models.items():
    # Train the model
    model.fit(X_train, y_train)
    
    # Make predictions
    y_pred = model.predict(X_test)
    
    # Calculate accuracy
    accuracy = accuracy_score(y_test, y_pred)
    
    # Append to the list
    model_names.append(model_name)
    model_accuracies.append(accuracy)

# Plot the comparison graph
plt.figure(figsize=(10, 6))
plt.barh(model_names, model_accuracies, color=['blue', 'green', 'red', 'purple', 'orange'])
plt.xlabel('Accuracy')
plt.title('Model Comparison: Accuracy of Different Models')
plt.xlim(0, 1)  # Set x-axis limit from 0 to 1
plt.grid(True, axis='x', linestyle='--', alpha=0.7)

# Display the accuracy on the bars
for i, v in enumerate(model_accuracies):
    plt.text(v + 0.02, i, f"{v:.4f}", va='center', fontweight='bold')
plt.savefig('model_comparison_plot.png', bbox_inches='tight', format='png')
plt.show()
```
